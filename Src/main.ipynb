# Simple Linear Regression

Code Implementation

```python
def predict(x, w, b):
    return w*x +b

def compute_loss(y_hat, y):
    loss = (y_hat - y)**2
    return loss

def compute_gradient(y_hat, y, x):
    dw = 2*x*(y_hat - y)
    db = 2*(y_hat - y)
    return dw, db

def update_weight(w, b, dw, db, lr):
    new_w = w - lr*dw
    new_b = b - lr*db
    return new_w, new_b
```

Test function

```python
#Befor training
x = 5
y = 66

b = 5
w = 10 
lr = 0.01
```

```python
y_hat = predict(x, w, b)
y_hat
```

```python
loss = compute_loss(y_hat, y)
loss
```

```python
dw, db = compute_gradient(y_hat, y, x)
dw, db
```

```python
new_w, new_b = update_weight(w, b, dw, db, lr)
new_w, new_b
```

```python
#Training
X = [3, 4, 5, 6]
y = [60, 55, 66, 93]

epochs = 5

b = 5
w = 10
lr = 0.001

for epoch in range(epochs):
    for idx in range(len(X)):
        x = X[idx]
        y_hat = predict(x, w, b)

        loss = compute_loss(y_hat, y[idx])
        dw, db = compute_gradient(y_hat, y[idx], x)
        w, b = update_weight(w, b, dw, db, lr)

        print(f'Epoch: {epoch}, Loss: {loss}, w: {w}, b: {b}')
```
